```python
import numpy as np
import matplotlib.pyplot as plt
import random
from scipy import stats
```

```python
s = [i for i in range(100) for _ in range (5)]
r = [0.125*i+random.uniform(-0.5, 0.5) for i in range(20) for _ in range(25)]
x = np.linspace(0, 100, 100)
```

```python
srx = np.std(r)
ssy = np.std(s)
rxbar = np.mean(r)
sybar = np.mean(s)
rxdevs = srx-rxbar
sydevs = ssy-sybar
rsxy = np.sum(rxdevs*sydevs)/np.sqrt(np.sum(rxdevs**2)*np.sum(sydevs**2))
b = rsxy*srx/ssy
a = 0
```

```python
plt.style.use("seaborn")
plt.title('Difference Stringency Index vs. Difference in Basic Reproduction Number for a sample of 100 ARs')
plt.xlabel('Difference in Stringency Index for Two Points in Time (Max - Min)')
plt.ylabel('Basic Reproductive Number')
plt.xlim([0, 100])
plt.scatter(s, r, s=20)
plt.plot(x, b*x+a, color="black")
plt.savefig('a.png')
plt.show()
```

```python
rptx = np.repeat(x, 5)
r_hat = b*rptx+a
resid = r - r_hat
print(resid[:20])
```

```python
plt.style.use("seaborn")
plt.title('Difference Stringency Index vs. Residuals')
plt.xlabel('Difference in Stringency Index for Two Points in Time (Max - Min)')
plt.ylabel('Residuals (observed - predicted) Basic Reproductive Number')
plt.xlim([0, 100])
plt.ylim([-0.75, 0.5])
plt.scatter(s, resid, s=20)
plt.savefig('c.png')
plt.show()
```

```python
nppresid = stats.probplot(resid, plot=plt)
plt.style.use("seaborn")
plt.title('Normal Probability Plot for Residuals')
plt.xlabel('Residuals (observed - predicted) Basic Reproductive Number')
plt.ylabel('Expected Z Score')
plt.savefig('d.png')
plt.show()
```
